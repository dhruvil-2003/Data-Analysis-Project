#### Duplicate data

```python
print("Total duplicate values = ",data.duplicated().sum())
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().any()
```

#### Courses per subject

```python
data["subject"].value_counts()
```

```python
a = sns.countplot(x = "subject",data = data,width = 0.8,hue = "subject",palette="dark")
plt.xlabel("Subjects",fontsize = 11)
plt.ylabel("Number of courses per subject",fontsize = 11)
plt.xticks(rotation = 30)
for x in a.containers:
    a.bar_label(x)
plt.show()
```

#### Levels for which Udemy provide courses

```python
data["level"].value_counts()
```

```python
a = sns.countplot(x = data["level"],hue = width = 0.8,palette="dark")
plt.xlabel("Level",fontsize = 11)
plt.ylabel("Number of courses per level",fontsize = 11)
plt.xticks(rotation = 30)
for x in a.containers:
    a.bar_label(x)
plt.show()
```

#### Count of Paid and Free Courses

```python
data["is_paid"].value_counts()
```

```python
y = sns.countplot(x = "is_paid",data = data,width = 0.4)
for x in y.containers:
    y.bar_label(x)
```

#### Course having more lectures(free or paid)

```python
data.groupby("is_paid")["num_lectures"].mean()
```

#### Courses having higher number of Subscribers

```python
b = sns.barplot(x = "is_paid",y = "num_subscribers",data = data,errorbar = None )
for x in b.containers:
    b.bar_label(x)
```

#### Courses having highest Subscribers

```python
data.groupby("level")["num_subscribers"].mean()
```

```python
y = sns.barplot(x = "level",y = "num_subscribers",data = data,hue = "is_paid",errorbar = None)
for x in y.containers:
    y.bar_label(x)
```

#### Most popular course title

```python
data[data["num_subscribers"].max() == data["num_subscribers"]]["course_title"]
```

```python
data[data["num_subscribers"].mean() == data["num_subscribers"]]["course_title"].head(10)
```

```python
top_10 = data.sort_values(by = "num_subscribers",ascending = False).head(10)
```

```python
sns.barplot(x = "num_subscribers",y = "course_title",hue = "course_title",data = top_10,legend = False)
```

#### Course having highest number of reviews

```python
z = data.sort_values(by = "num_reviews",ascending = False)
```

```python
plt.figure(figsize = (10,4))
sns.barplot(x = "subject",y = "num_reviews",hue = "subject",data = z)
```

#### Does price affect number of reviews

```python
plt.figure(figsize = (13,6))
sns.scatterplot(x = "price",y = "num_reviews",data = data)
```

#### Number of courses related to python

```python
len(data[data["course_title"].str.contains("python",case = False)])
```

#### 10 most popular python courses as per number of subscribers

```python
a = data[data["course_title"].str.contains("python",case = False)].sort_values(by = "num_subscribers",ascending = False).head(10)
```

```python
plt.figure(figsize = (13,6))
sns.barplot(y = "course_title",x = "num_subscribers",hue = "num_subscribers",palette="dark",data = a,orient = "h")
```
