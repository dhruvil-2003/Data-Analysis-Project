#### Healthcare Access V/S Mortality Rate

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'Healthcare Access (%)',y = 'Mortality Rate (%)',data = data,hue = 'Disease Category')
plt.title("Healthcare Access v/s Mortality rate")
plt.xlabel("Healthcare Access(%)")
plt.ylabel("Mortality Rate(%)")
plt.show()
```

#### Top 5 most prevalence diseases

```python
data.head()
```

#### Which disease have higher Prevalence rate

```python
a = data.groupby('Disease Name')['Prevalence Rate (%)'].sum().sort_values(ascending = False).head()
a
```

```python
q = sns.barplot(a)
for x in q.containers:
    q.bar_label(x)
```

##### Conclusion : Cholera have higher prevalence rate.

#### Correlation between disease prevalence and socio-economic factors like income,education and urbanization

```python
health_corr=data[['Prevalence Rate (%)','Per Capita Income (USD)','Education Index','Urbanization Rate (%)']]

heatmap = sns.heatmap(health_corr.corr(numeric_only=True), annot=True)
```

#### Country with healthcare access and hospital beds below average

#### Countries with low Healthcare Access

```python
plt.figure(figsize = (10,6))
sns.barplot(y = 'Country', x = 'Healthcare Access (%)',hue= 'Country',data = priority_countries.sort_values('Healthcare Access (%)'),palette='Reds')
plt.title('Countries with low Healthcare Access')
plt.xlabel('Healthcare Access(%)')
plt.ylabel('Countries')
plt.show()
```
