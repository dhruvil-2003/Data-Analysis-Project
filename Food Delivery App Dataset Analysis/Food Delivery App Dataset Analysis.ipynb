#### Checking how many restaurents are categorized into different listed_in(type)

```python
data[data['listed_in(type)'] == 'Dining'].count()
```

```python
a = sns.countplot(x = "listed_in(type)",data = data,hue = "listed_in(type)")
for x in a.containers:
    a.bar_label(x)
```

##### Conclusion: The majority of the restaurants fall into the dining category.

#### People prefer which type of listed_in(type)

```python
grouped_data = data.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
```

```python
plt.plot(result,c='orange')
plt.show()
```

##### Conclusion: Dining restaurants are preferred by a larger number of individuals.

#### Restaurents with maximum votes

```python
data.groupby('name')['votes'].max().sort_values(ascending = False).head(1)
```

#### How many ordered online or offline

```python
z = sns.countplot(x = 'online_order',data = data,hue = 'online_order')
for x in z.containers:
    z.bar_label(x)
```

##### Conclusion: This suggests that a majority of the restaurants do not accept online orders.

#### People provided which ratings the most

```python
r = sns.histplot(x = 'rate',data = data,hue = 'rate',legend = False)
for x in r.containers:
    r.bar_label(x)
plt.xticks(rotation = 60)
plt.show()
```

##### Conclusion: The majority of restaurants received ratings ranging from 3.5 to 4.

#### How many couples spended average money on meals

```python
v = sns.countplot(x = 'approx_cost(for two people)' , data = data,hue = 'approx_cost(for two people)',legend = False)
plt.xticks(rotation = 40)
for x in v.containers:
    v.bar_label(x)
plt.show()
```

##### Conclusion: The majority of couples prefer restaurants with an approximate cost of 300 rupees.

#### Which order online or offline received maximum voting

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order',y = 'rate',data = data,hue = 'online_order')
plt.show()
```

##### Conclusion:Offline order received less ratings than online order

```python
pivot_table = data.pivot_table(index = 'listed_in(type)',columns = 'online_order',aggfunc = 'size',fill_value = 0)
sns.heatmap(pivot_table,annot = True,cmap = 'YlGnBu',fmt = 'd')
plt.title('Heatmap')
plt.xlabel('Online order')
plt.ylabel('Listed In (Type)')
plt.show()
```

#### Finding out the average cost per listed_in(type)

```python
s = sns.barplot(x = 'listed_in(type)' , y = 'approx_cost(for two people)', data = data,errorbar = None,hue = 'listed_in(type)')
for x in s.containers:
    s.bar_label(x)
```

#### How much money is earned by dining , cafes , other and buffet

```python
w = data.groupby('listed_in(type)')['approx_cost(for two people)'].sum().sort_values(ascending = False)
w
```

```python
q = sns.barplot(w,palette = 'gist_rainbow')
for x in q.containers:
    q.bar_label(x)
    
```

##### Conclusion : We made the conclusion that dining had huge earnings followed by Cafes , other and Buffet had least earnings
