#### Most profitable branch as per gross income

```python
q = sns.barplot(x = "Branch",y = "gross income",hue = "City",data = data,errorbar = None)
for x in q.containers:
    q.bar_label(x)
```

#### Relation between gender and gross income

```python
sns.boxplot(x = "Gender",y = "gross income",hue = "Customer type",data = data)
```

#### Product line that generates more income

```python
w = sns.barplot(x = "Product line",y = "gross income",data = data,hue = "Product line",errorbar = None)
plt.xticks(rotation = 30)
for x in w.containers:
    w.bar_label(x)
plt.show()
```

#### Highest Unit price in the product line

```python
e = sns.barplot(x = "Product line",y = "Unit price",data = data,hue = "Product line",errorbar = None)
plt.xticks(rotation = 50)
for x in e.containers:
    e.bar_label(x)
plt.show()    
```

#### Product line which is purchased in highest quantity

```python
(data.groupby("Product line")["Quantity"].sum().sort_values(ascending = False)).plot(kind = "bar")
```

#### Display daily sales by day of the week

```python
dw_mapping = {
    0:"Mon",
    1:"Tue",
    2:"Wed",
    3:"Thur",
    4:"Fri",
    5:"Sat",
    6:"Sun"
}
```

```python
data["day_of_week"] = data["Date"].dt.dayofweek.map(dw_mapping)
```

```python
(data["day_of_week"].value_counts()).plot(kind = "bar")
```

#### What will be the highest months for sales

```python
month_mapping = {
    1:"Jan",
    2:"Feb",
    3:"March"
}
```

```python
data["month"] = data["Date"].dt.month.map(month_mapping)
```

```python
sns.barplot(data["month"].value_counts())
```
